```python
#librerias necesarias
import random
import numpy as np
import matplotlib.pyplot as plt
import statistics 

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
#Posibles números ruleta
lista_posible = list(range(37))
print('Posibles valores:',lista_posible,'\n',statistics.mean(lista_posible),statistics.pvariance(lista_posible))
```

```python
#Vamos a tirar la ruleta muchas veces para ver qué valores salen
n = 10000
valores_obtenidos = []
for i in range(n):
    valores_obtenidos.append(random.randrange(37))
#print('Resultado luego de ',n,' iteraciones: ',valores_obtenidos,'\n')

#Esperanza matemática
esperanza = statistics.mean(valores_obtenidos)
varianza = statistics.pvariance(valores_obtenidos)
print('ESPERANZA MATEMÁTICA: ',esperanza,'VARIANZA: ',varianza)
```

```python
#Gráfica 1
#Histograma
plt.hist(valores_obtenidos, bins=37, alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.title('HISTOGRAMA')  # Colocamos el título
plt.xlabel('valores ruleta') # Nombre x
plt.ylabel('frecuencia absoluta') # Nombre y
plt.show()
plt.clf()
```

```python
#Gráfica 2
#Frecuencia relativa en n tiradas para un valor particular
#Tomamos al azar un valor
valor = random.randrange(37)
#Calculamos la frecuencia relativa en n tiradas
lista_f_relativa = []
num_apariciones = 0
for i in range(1,n):
    numero_sale = random.randrange(37)
    #tirada,frecuencia relativa
    if valor == numero_sale:
        num_apariciones +=1
    f_r = num_apariciones/i
    lista_f_relativa.append([i,f_r])
    
x, y = zip(*[i for i in lista_f_relativa])
#print(x,y)
plt.title('Frecuencia relativa del número '+ str(valor))  # Colocamos el título
plt.plot(x, y,'ro-', markersize=0.5, lw=0.5)
plt.plot([1/37 for i in range(n)], linestyle='dashed', color='blue')
plt.grid(True)
plt.xlabel('Tiradas')
plt.ylabel('Fr')
plt.show()
```

```python
#Gráfica 3
#Esperanza matemática en n tiradas

lista_esperanza = []
for i in range(n):
    #Esperanza matemática
    #valores_obtenidos[:i+1] voy agregando de a 1 los valores obtenidos en las n tiradas
    esperanza_i = statistics.mean(valores_obtenidos[:i+1])
    lista_esperanza.append([i,esperanza_i])

x, y = zip(*[m for m in lista_esperanza])

plt.title('Esperanza en {} tiradas'.format(n))  # Colocamos el título
plt.plot(x, y,'ro-', markersize=0.5, lw=0.5)
plt.plot([esperanza for i in range(n)], linestyle='dashed', color='blue')
plt.grid(True)
plt.xlabel('Tiradas')
plt.ylabel('Esperanza')
plt.show()
```

```python
#Gráfica 4
#Esperanza de la esperanza
#Vamos a calcular la esperanza n veces
lista_esperanza_esperanza = []
for e in range(10): #Comparo la esperanza de la esperanza 10 veces
    valores_obtenidos = []
    lista_esperanza = []
    for i in range(n):
        valores_obtenidos.append(random.randrange(37))
        #Esperanza matemática
        esperanza = statistics.mean(valores_obtenidos)
        lista_esperanza.append([i,esperanza])
    lista_esperanza_esperanza.append(lista_esperanza)
```

```python
#Graficamos la esperanza de la esperanza
plt.title('Esperanza de la esperanza en {} tiradas'.format(n))  # Colocamos el título
for e in lista_esperanza_esperanza:
    x,y = zip(*[m for m in e])
    plt.plot(x, y,'ro-', markersize=0.5, lw=0.5)
    
plt.plot([esperanza for i in range(n)], linestyle='dashed', color='blue')

plt.grid(True)
plt.xlabel('Tiradas')
plt.ylabel('Esperanza de la esperanza')
plt.show()
```

```python
#Gráfica 5
#Varianza matemática en n tiradas

lista_varianza = []
for i in range(n):
    #Varianza matemática
    varianza_i = statistics.pvariance(valores_obtenidos[:i+1])
    lista_varianza.append([i,varianza_i])

x, y = zip(*[m for m in lista_varianza])

plt.title('Varianza en {} tiradas'.format(n))  # Colocamos el título
plt.plot(x, y,'ro-', markersize=0.5, lw=0.5)
plt.plot([varianza for i in range(n)], linestyle='dashed', color='blue')
plt.grid(True)
plt.xlabel('Tiradas')
plt.ylabel('Varianza')
plt.show()
```
